## Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/bikeshare.csv'
bikes = pd.read_csv(url)

bikes.head()
```

**Plot Data**

```python
bikes.plot(kind='scatter', x='temp', y='count', alpha=0.2)
```

```python
import seaborn as sns
sns.lmplot(x='temp', y='count', data=bikes, aspect=1.5, scatter_kws={'alpha': 0.2})
```

```python
# Correlation between count and temp
bikes[['count', 'temp']].corr()
```

**Linear Model**

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
X = bikes[['temp']]
y = bikes['count']
```

For laziness we did not split in training and test set

```python
linreg = LinearRegression()
linreg.fit(X, y)
```

```python
print('Intercept: {}'.format(linreg.intercept_))
print('Slope: {}'.format(linreg.coef_))
```

```python
# Make prediction
linreg.predict(20)
```

**Adding more predictors**

```python
feature_cols = ['temp', 'season', 'weather', 'humidity']
sns.pairplot(bikes, x_vars=feature_cols, y_vars='count', kind='reg')
```

```python
X = bikes[feature_cols]
y = bikes['count']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

We perform prediction on the test set

```python
y_pred = linreg.predict(X_test)
```

```python
print('RSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

**Null Model**

```python
# Let's suppose that the null model predict always the mean value of rented bikes
average_bike_rental = bikes['count'].mean()
null_model = [average_bike_rental] * bikes.shape[0]
```

```python
# RSE of the null model
print('RSE:', np.sqrt(metrics.mean_squared_error(y, null_model)))
```
